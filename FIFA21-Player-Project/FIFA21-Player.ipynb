```python
import pandas as pd
df = pd.read_csv(r"fifa21_raw_data.csv")
df
```

```python
df.info()
```

Convert the height and weight columns to numerical forms

```python
def height_to_meters(height):
    feet_inch = height.split("'")
    return round(int(feet_inch[0])*0.3048 + int(feet_inch[1][0])*0.0254, 2)

df['Height'] = df['Height'].apply(height_to_meters)
df['Height']
```

```python
def weight_to_kg(weight):
    return round(int(weight[:-3])*0.453592,2)

df['Weight'] = df['Weight'].apply(weight_to_kg)
```

```python
df['Weight']
```

Clean and transform the value, wage and release clause columns into columns of integers.

```python
def to_numeric(num):
    match num[-1]:
        case "M":
            return int(float(num[1:-1])*10**6)
        case "K":
            return int(float(num[1:-1])*10**3)
        case _:
            return int(num[1:])

df['Release Clause'] = df['Release Clause'].apply(to_numeric)
df['Value'] = df['Value'].apply(to_numeric)
df['Wage'] = df['Wage'].apply(to_numeric)

df[['Release Clause', 'Wage', 'Value']]

```

Separate the Team & Contract column into separate team and contract columns

```python
df[['Team', 'Contract']] = df['Team & Contract'].str.strip().str.split("\n", expand=True)
df[['Team', 'Contract']]
```

```python
df = df.drop(columns=['Team & Contract'])
df.head()
```

Separate the joined column into year, month and day columns.

```python

df[['Joined Month', 'Joined Day', 'Joined Year']] = df['Joined'].str.replace(',', '').str.split(" ", expand=True)
df[['Joined Month', 'Joined Day', 'Joined Year']] 
```

Remove the unnecessary newline characters from the Hits columns and convert it to integer values.

```python
def hits_to_numeric(num):
    return int(num) if 'K' not in num else int(float(num[:-1])*1000)

df['Hits'] = df['Hits'].astype(str).str.replace("\n", "").apply(hits_to_numeric)
df['Hits']
```

Check which players have been playing at a club for more than 10 years!

```python
df['Joined'] = pd.to_datetime(df['Joined'])
df['Joined']
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta

def more_than_10_years(date):
    return relativedelta(datetime.now(), date).years > 10

df[df['Joined'].apply(more_than_10_years)]
```

```python
import pandas as pd
df = pd.read_csv(r'FIFA21_final.csv')
```

```python
df.head()
```

This scatter plot shows the relationship between the wage and the expected offensive value added (xOVA) of the players. The plot shows that the wage is proportional to xOVA.

```python
df = df.rename(columns={'↓OVA': 'xOVA'})
sns.scatterplot(data=df, x='xOVA', y='Wage')
plt.title('xOVA vs. Wage')
```

Let's also look at the top paid players.

```python
df.nlargest(10,columns=['Wage'], keep='all')[['Name', 'Team', 'Wage', 'xOVA']]
```

This scatter plot also shows the relationship between the xOVA and the value of the players. The plot shows that the value increases exponentially with xOVA, which reflects the market demand and preference for players who can score goals and create chances.

```python
sns.regplot(data=df, x='xOVA', y='Value')
plt.title('xOVA vs. Value')
```

This histogram shows the distribution of wage among the players. The histogram is unevenly distributed, with most of the players receiving low wages and only a few players receiving very high wages. This indicates a high level of inequality and imbalance in the distribution of income among the players.

```python
values, bins, bars = plt.hist(x='Wage', bins=30, data=df[df['Wage'] != 0])
plt.bar_label(bars, fontsize=12, color='navy')
plt.title('Wage Distribution')
plt.xlabel('Wage')
```

This bar plot shows the top 10 nationalities that have the highest average xOVA among their players. The plot shows that most of these nationalities are from Europe and South America, such as Portugal, Argentina, Brazil, France, etc. This suggests that these regions have more talented and skilled players than other regions.

```python
df_avg_xova = df.groupby('Nationality')[['xOVA']].mean().nlargest(10,columns='xOVA').reset_index()
ax = sns.barplot(data=df_avg_xova,y='Nationality', x='xOVA')
ax.bar_label(ax.containers[0])
plt.title('xOVA vs. Nationality')
```

This line plot shows the oldest teams in terms of average age of their players. The plot shows that most of these teams are from lower leagues or less competitive leagues, such as FC Tokyo, Cork City, FC Midtjylland, etc. This implies that these teams have less resources or opportunities to recruit younger and more promising players.

```python
oldest_team_df = df.groupby('Team')[['Age']].mean().nlargest(10, columns='Age').reset_index()
sns.lineplot(data=oldest_team_df, x='Age', y='Team')
plt.title('Team Age')
for x,y in zip(oldest_team_df['Age'], oldest_team_df['Team']):
    plt.text(x = x + 0.1, y = y, s = '{:.2f}'.format(x), color = 'Navy')
```
